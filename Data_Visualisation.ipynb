```python
df1=pd.read_csv("/data/Titanic.csv")
df2=pd.read_csv("/data/house_prices.csv")
```

<h1><em>Histograms</em></h1>

```python
plt.figure(figsize=(10, 6))
plt.hist(df1['Age'], bins=8, edgecolor='k', alpha=0.7)
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
#Bins is the number of divisions of the histogram
```

<h1><em>Box Plots</em></h1>

<h3>The boxes depict the median range while the whiskers depict the minimum and maximum values in a certain range.</h3>

```python
plt.figure(figsize=(40, 12))
sns.boxplot(x='Age', y='Pclass', data=df1)
plt.title('Box Plot of Passenger Class by Age')
plt.xlabel('Age')
plt.ylabel('Passenger Class')
plt.grid(True)
plt.show()
```

<h1><em>Scatter Plots</em></h1>

<h3>Used when you want to detect trends like a positive or negative correlation</h3>

```python
plt.figure(figsize=(10, 6))
plt.scatter(df1['Age'], df1['Fare'], alpha=0.7)
plt.title('Scatter Plot of Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

<h1><em>Pair Plots</em></h1>

```python
pd.option_context('mode.use_inf_as_na', True)
pair_plot=sns.pairplot(df2[['YrSold', 'LotArea', 'SalePrice']],height=3.5)
pair_plot.fig.suptitle('Pair Plot of YrSold, LotArea, and SalePrice', y=1.02)  
plt.show()
```

<h1><em>Heatmaps</em></h1>

```python
df1['Sex_numerical']=df1['Sex'].map({"male":1,"female":0})
corr = df1[['Pclass', 'Sex_numerical', 'Survived']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

<h3>We see there is medium negative correlation between the Passenger class and the chance to survive implying that higher passenger classes tend to have a higher chance of surviving and a high negative correlation between Sex and the chance of survival implying that females had a higher survival rate compared to males.</h3>

<h1><em>Bar Charts</em></h1>

```python
plt.figure(figsize=(6, 6))
sns.countplot(x='Sex', data=df1)
plt.title('Bar Chart of Sex Count')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

<h1><em>Violin Plots</em></h1>

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='YrSold', y='SalePrice', data=df2)
plt.title('Violin Plot of Salary by Department')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

<h1><em>Line Plots</em></h1>

```python
plt.figure(figsize=(20, 6))
plt.plot(df2['YearBuilt'], df2['SalePrice'], marker='x', linestyle='', color='b')
plt.title('Line Plot of sale prices for the year the house was built')
plt.xlabel('Time')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```
