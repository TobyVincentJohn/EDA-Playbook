```python
import pandas as pd
df=pd.read_csv("/data/Titanic.csv")
```

<h1><em>Central Tendency Measures</em></h1>

```python
mean_age = df['Age'].mean()
median_age = df['Age'].median()
mode_age = df['Age'].mode()[0]

mean_fare = df['Fare'].mean()
median_fare = df['Fare'].median()
mode_fare = df['Fare'].mode()[0]

print("Mean Age:", mean_age)
print("Median Age:", median_age)
print("Mode Age:", mode_age)

print("\nMean Fare:", mean_fare)
print("Median Fare:", median_fare)
print("Mode Fare:", mode_fare)
```

<h1><em>Dispersion Measures</em></h1>

<h3>Variance measures the average squared deviation of each data point from the mean of the dataset. It quantifies the degree of spread in the data.</h3>

<h3>Standard deviation is the square root of the variance. It provides a measure of dispersion in the same units as the data, making it more interpretable compared to variance.</h3>

```python
std_dev_age = df['Age'].std()
variance_age = df['Age'].var()

std_dev_fare = df['Fare'].std()
variance_fare = df['Fare'].var()

print("\nStandard Deviation of Age:", std_dev_age)
print("Variance of Age:", variance_age)

print("\nStandard Deviation of Fare:", std_dev_fare)
print("Variance of Fare:", variance_fare)
```

<h1><em>Shape of Data Distribution</em></h1>

<h3>Skewness measures asymmetry of the data distribution, it quantifies the extent to which a distribution deviates from a symmetric bell curve</h3>

**Positive Skewness**: If the skewness is positive, the distribution has a longer tail on the right side. This indicates that the majority of data points are concentrated on the left with a few larger values stretching out the right tail.

**Negative Skewness**: If the skewness is negative, the distribution has a longer tail on the left side. This indicates that the majority of data points are concentrated on the right with a few smaller values stretching out the left tail.

**Zero Skewness**: A skewness of zero indicates a perfectly symmetric distribution

```python
skewness_age = df['Age'].skew()
skewness_fare = df['Fare'].skew()

print("\nSkewness of Age:", skewness_age)
print("\nSkewness of fare:", skewness_fare)
```

<h3>Kurtosis measures the "tailedness" or the extremity of outliers in the data distribution. It indicates the height and sharpness of the distribution's peak.</h3>

**Positive Kurtosis (Leptokurtic)**: Indicates that the distribution has heavy tails and a sharp peak compared to a normal distribution. This means there are more extreme outliers.

**Negative Kurtosis (Platykurtic)**: Indicates that the distribution has lighter tails and a flatter peak compared to a normal distribution. This means fewer extreme outliers.

**Zero Kurtosis (Mesokurtic)**: A kurtosis of zero indicates a normal distribution with a standard peak and tails.

```python
kurtosis_age = df['Age'].kurt()
kurtosis_fare = df['Fare'].kurt()

print("Kurtosis of Age:", kurtosis_age)
print("Kurtosis of fare:", kurtosis_fare)
```

<h1><em>Percentiles and Quartiles</em></h1>

```python
age_25th_percentile = df['Age'].quantile(0.25)
age_50th_percentile = df['Age'].quantile(0.5)  # Equivalent to the median
age_75th_percentile = df['Age'].quantile(0.75)

fare_25th_percentile = df['Fare'].quantile(0.25)
fare_50th_percentile = df['Fare'].quantile(0.5)  # Equivalent to the median
fare_75th_percentile = df['Fare'].quantile(0.75)

print("\nAge Percentiles:")
print("25th Percentile:", age_25th_percentile)
print("50th Percentile (Median):", age_50th_percentile)
print("75th Percentile:", age_75th_percentile)

print("\nFare Percentiles:")
print("25th Percentile:", fare_25th_percentile)
print("50th Percentile (Median):", fare_50th_percentile)
print("75th Percentile:", fare_75th_percentile)
```
